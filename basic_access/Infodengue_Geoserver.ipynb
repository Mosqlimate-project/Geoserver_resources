```python
import geopandas as gpd
import requests
import geojson
from pyproj import CRS
from owslib.wfs import WebFeatureService
```

## WFS server do Infodengue

```python
url = "https://info.dengue.mat.br/geoserver/wfs"
```

Primeiro precisamos estabelecer os parâmetros do Download.

```python
params = dict(
    service="WFS",
    version="2.0.0",
    request="GetFeature",
    typeName="BR_distritos_CD2022",
    outputFormat="json",
)
```

Então fazemos a requisição, combinando a url com os parâmetros:

```python
r = requests.get(url, params=params)
```

O download dos dados só ocorre quando rodamos a linha abaixo:

```python
distritos = gpd.GeoDataFrame.from_features(geojson.loads(r.content), crs="EPSG:4674")
```

```python
distritos.plot()
```
